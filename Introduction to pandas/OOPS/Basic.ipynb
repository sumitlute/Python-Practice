```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
    def get_grade(self):
        return f"{self.name} is in grade {self.grade}."
```

```python
Sumit = Student("Sumit", 10)
print(Sumit.get_grade())
```

```python
Sayli = Student("Sayli", 12)
print(Sayli.get_grade())
```

```python
class St:
    def __init__(self,name,course,number,attendence):
        self.name = name
        self.course = course
        self.number = number
        self.attendence = attendence
    def topic(self):
        return f"{self.name} {self.attendence} today for {self.course} and can be connected at {self.number}"
```

```python
class BankAccount:

    def __init__(self,customer_name,balance):
        self.customer_name = customer_name
        self.balance = balance
  
    def deposit(self, amount):
        self.balance += amount
        print(f"₹{amount} deposited.")

    def withdraw(self, amount):
        if amount <= self.amount:
            self.balance -= amount
            print(f"₹{amount} withdrawn.")
        else:
            print("Insufficant Balance")

def check_balance(self):
    print(f"Current Balance: ₹{self.balance}")
    
```

```python
account = BankAccount("Sumit", 10000)
account.deposit(5000)
account.withdraw(3000)
account.check_balance()
```
