```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
    def get_grade(self):
        return f"{self.name} is in grade {self.grade}."
```

```python
class St:
    def __init__(self,name,course,number,attendence):
        self.name = name
        self.course = course
        self.number = number
        self.attendence = attendence
    def topic(self):
        return f"{self.name} {self.attendence} today for {self.course} and can be connected at {self.number}"
```

```python
Sumit = St("sumit", "absent", "data analytics", 8530693118)
print(Sumit.topic())
```

```python
class BankAccount:

    def __init__(self,customer_name,balance):
        self.customer_name = customer_name
        self.balance = balance
  
    def deposit(self, amount):
        self.balance += amount
        print(f"₹{amount} deposited.")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"₹{amount} withdrawn.")
        else:
            print("Insufficant Balance")

    def check_balance(self):
        print(f"Current Balance: ₹{self.balance}")
    
```

```python
account = BankAccount("Sumit", 10000)
account.deposit(5000)
account.withdraw(3000)
account.check_balance()
```

```python
class Product:
    def __init__(self,name,price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price:", self.price)

product1 = Product("Laptop", 50000)

product1.display()
```

```python
class Patitent:
    
    def __init__(self,patitent_id,name):
        self.patitent_id = patitent_id
        self.name = name

    def display(self):
        print("Patitent_Id:", self.patitent_id)
        print("Name:", self.name)

patitent1 = Patitent("P101", "Sumit")

patitent1.display()
```

```python
class Machine:
    
    def __init__(self,machine_id,status):
        self.machine_id = machine_id
        self.status = status

    def start(self):
        self.status = "Running"
        print("Machine Started")

    def stop(self):
        self.status = "Stopped"
        print("Machine Stopped")

    def display(self):
        print(self.machine_id, self.status)

machine1 = Machine("M101", "Stopped")

machine1.start()
machine1.display()
```
